```python
def FacSimple(n):
    factorial = 1
    i = 1
    while i <= n:
        factorial *= i
        i += 1

    return factorial

FacSimple(5)
```

```python
def factorial(n):
    result = 1
    for i in range(2, n+1):
        result *= i
    return result

print(FacSimple(5))
print("""
    Оцінка асимптотичної складності:

    Часова складність:
    - Алгоритм використовує один цикл for, що ітерується від 2 до n включно.
    - Кількість ітерацій циклу є пропорційною n.
    - Кожна ітерація виконує постійну кількість операцій (множення та присвоєння).
    - Отже, часова складність алгоритму є O(n).

    Просторова складність:
    - Алгоритм використовує постійну кількість додаткової пам'яті (змінна result та змінна циклу i).
    - Отже, просторова складність алгоритму є O(1).
    """)
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    if n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(30), """
    Рекурсивна функція для обчислення n-го числа Фібоначчі.

    Алгоритм використовує рекурсію, що призводить до експоненційного
    зростання кількості операцій при збільшенні n.

    Складність:
    - Час: O(2^n) в найгіршому випадку. Кожен виклик функції генерує
           два додаткові виклики, що призводить до експоненційного
           зростання кількості операцій.
    - Простір: O(n) в найгіршому випадку. Глибина стеку рекурсії зростає
              лінійно зі збільшенням n.""")
     
```

```python
def merge(left, right):
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result
def mergesort(list):
    if len(list) < 2:
        return list
    middle = len(list) // 2
    left = mergesort(list[:middle])
    right = mergesort(list[middle:])
    return merge(left, right)
a = [6, 5, 3, 1, 8, 7, 2, 4]
print(merge([2, 4, 5, 7],[1, 2, 3, 6]),"\n",mergesort(a),"""
Складність:
    - Час: O(n log n), де n - кількість елементів у вхідному масиві.
           Час сортування обумовлений рекурсивним розбиттям масиву на підмасиви,
           їх сортуванням і об'єднанням, що займає O(n log n) часу.
    - Простір: O(n), де n - кількість елементів у вхідному масиві.
              Додатковий простір для зберігання підмасивів під час рекурсивних викликів.""")
```
